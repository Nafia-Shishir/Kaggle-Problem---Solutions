# Data processing

```python
train_data['LotArea'] = np.log(train_data['LotArea']+1)
train_data['MasVnrArea'] = np.log(train_data['MasVnrArea']+1)
train_data['PoolArea'] = np.log(train_data['PoolArea']+1)
train_data['MiscVal'] = np.log(train_data['MiscVal']+1)
```

# Feature Engineering and data cleaning

```python
train_data=train_data.join(pd.get_dummies(data.MSZoning)).drop(['MSZoning'],axis=1)
```

```python
train_data=train_data.join(pd.get_dummies(data.LotShape)).drop(['LotShape'],axis=1)
```

```python
train_data=train_data.join(pd.get_dummies(data.LandContour)).drop(['LandContour'],axis=1)
```

```python
train_data=train_data.join(pd.get_dummies(data.Utilities)).drop(['Utilities'],axis=1)
```

```python
train_data=train_data.join(pd.get_dummies(data.LotConfig)).drop(['LotConfig'],axis=1)
```

```python
train_data=train_data.join(pd.get_dummies(data.Condition1)).drop(['Condition1'],axis=1)
```

```python
train_data=train_data.join(pd.get_dummies(data.BldgType)).drop(['BldgType'],axis=1)
```

```python
train_data=train_data.join(pd.get_dummies(data.ExterCond)).drop(['ExterCond'],axis=1)
```

```python
train_data=train_data.join(pd.get_dummies(data.Functional)).drop(['Functional'],axis=1)
```

```python
train_data=train_data.join(pd.get_dummies(data.SaleCondition)).drop(['SaleCondition'],axis=1)
```

```python
plt.figure(figsize=(60,50))
sns.heatmap(train_data.corr(), annot=True, cmap="YlGnBu")
```

```python
test_data = X_test.join(y_test)

test_data['LotArea'] = np.log(test_data['LotArea']+1)
test_data['MasVnrArea'] = np.log(test_data['MasVnrArea']+1)
test_data['PoolArea'] = np.log(test_data['PoolArea']+1)
test_data['MiscVal'] = np.log(test_data['MiscVal']+1)

test_data=test_data.join(pd.get_dummies(data.MSZoning)).drop(['MSZoning'],axis=1)
test_data=test_data.join(pd.get_dummies(data.LotShape)).drop(['LotShape'],axis=1)
test_data=test_data.join(pd.get_dummies(data.LandContour)).drop(['LandContour'],axis=1)
test_data=test_data.join(pd.get_dummies(data.Utilities)).drop(['Utilities'],axis=1)
test_data=test_data.join(pd.get_dummies(data.LotConfig)).drop(['LotConfig'],axis=1)
test_data=test_data.join(pd.get_dummies(data.Condition1)).drop(['Condition1'],axis=1)
test_data=test_data.join(pd.get_dummies(data.BldgType)).drop(['BldgType'],axis=1)
test_data=test_data.join(pd.get_dummies(data.ExterCond)).drop(['ExterCond'],axis=1)
test_data=test_data.join(pd.get_dummies(data.Functional)).drop(['Functional'],axis=1)
test_data=test_data.join(pd.get_dummies(data.SaleCondition)).drop(['SaleCondition'],axis=1)
```
